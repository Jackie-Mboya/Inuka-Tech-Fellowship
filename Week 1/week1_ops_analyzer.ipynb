### Operational KPI Analyzer: Healthcare Facility Efficiency
#### Target Audience: Junior Operations Analyst

This notebook is designed to automate how we track operational efficiency across different healthcare sectors (e.g., hospital clinics or wards). 



**Logic Breakdown:**
1. **Mathematical KPI Calculation:** We determine efficiency by dividing actual patient care output by the target baseline.
2. **Conditional Logic (Status Thresholds):** We map percentages to qualitative health statuses (`Normal`, `Warning`, or `Critical`).
3. **Structured Data Modeling:** We use a collection of key-value records to cleanly house each facility's raw metrics.
4. **Automation Loop:** We cycle through each facility automatically to generate a standardized report.

#### 1. KPI Function: Write a function calculate_efficiency(actual_output, target_output) that returns the efficiency percentage.

```python
def calculate_efficiency(actual_output, target_output):
    """
    Calculates the operational efficiency as a percentage.
    Formula: (Actual Output / Target Output) * 180
    """
    if target_output == 0:
        return 0.0
    efficiency = (actual_output / target_output) * 100
    return round(efficiency, 2)
```

### The KPI Calculation Function

First, we need a standard, reusable way to calculate efficiency. Instead of writing out the math for every single ward manually, we package it inside a function called `calculate_efficiency`.

**Logic Breakdown:**
* **Inputs:** It accepts `actual_output` (the actual number of patients cared for) and `target_output` (the baseline target goal for that unit).
* **Edge Case Handling:** It checks if the target is zero to avoid a critical mathematical error (division by zero).
* **Output:** It divides the actual metrics by the target, multiplies by 100 to get a percentage, and rounds it to 2 decimal places for a clean presentation.

### 2. Status Classifier: Write a function get_operational_status(efficiency) that returns:

- "Critical" if efficiency < 70%
- "Warning" if efficiency is between 70% and 89%
- "Normal" if efficiency >= 90%

```python
def get_operational_status(efficiency):
    """
    Classifies operational health based on pre-defined performance brackets.
    """
    if efficiency < 70:
        return "Critical"
    elif 70 <= efficiency <= 89:
        return "Warning"
    else:
        return "Normal"
```

### The Operational Status Classifier Function

Now that we can calculate the exact percentage, we need a way to interpret what that number means operationally. We use a function called `get_operational_status` to sort our results into clear action categories.

**Logic Breakdown:**
* **Input:** It takes the efficiency percentage we just calculated.
* **Conditional Logic:** It uses an `if-elif-else` chain to test which performance bracket the percentage falls into:
  * **Critical:** Anything strictly below 70%.
  * **Warning:** Margins between 70% and 89% where operational cracks are starting to show.
  * **Normal:** Optimal performance at 90% or above.
* **Output:** It returns a simple text string indicating the status flag.

### Data Modeling & Iteration

Create a dictionary representing 3 different operational sites (you choose the industry: e.g., 3 Fuel Depots, 3 Hospital Wards, or 3 Retail Stores). Each site should have keys for name , actual_output , and target_output .

**The Logic**


Before we can automate the reporting, we need a clean, structured way to store our data. We are using a list of **dictionaries**. 

Think of a dictionary as a digital file card where every piece of information is explicitly labeled with a "key" (like `name`, `actual_output`, or `target_output`). Grouping these cards into a single list allows us to keep all our active hospital wards organized in one place.

```python
# Modeling data for 3 Hospital Wards
healthcare_sites = [
    {
        "name": "Maternal & Newborn Care Unit",
        "actual_output": 45,  # Patients fully processed/discharged seamlessly
        "target_output": 50
    },
    {
        "name": "Outpatient Emergency Clinic",
        "actual_output": 110,
        "target_output": 120
    },
    {
        "name": "Pediatric Specialized Ward",
        "actual_output": 28,
        "target_output": 45
    }
]
```

### 4. Iteration and Automated Reporting

This is where the automation happens! We use a `for` loop to step through our list of hospital units one by one. 

**Logic Breakdown:**
1. **The Loop:** The script automatically selects a unit's data card.
2. **Execution:** It feeds that unit's raw values directly into our `calculate_efficiency` and `get_operational_status` functions.
3. **Formatting:** It prints a neat, standardized visual text report with clear dividers. 
4. **Repeat:** It instantly moves to the next unit until every facility has been analyzed—meaning this code works just as fast whether we have 3 wards or 300!

```python
# Execution and reporting loop
print("==================================================")
print("       WEEKLY HEALTH FACILITY OPERATIONS REPORT   ")
print("==================================================\n")

for site in healthcare_sites:
    # Calculate efficiency
    eff_pct = calculate_efficiency(site["actual_output"], site["target_output"])
    
    # Categorize status
    status = get_operational_status(eff_pct)
    
    # Print formatted results
    print(f"Facility Name : {site['name']}")
    print(f"Target Output : {site['target_output']} patients")
    print(f"Actual Output : {site['actual_output']} patients")
    print(f"Efficiency    : {eff_pct}%")
    print(f"Status        : [{status}]")
    print("-" * 50)
```
